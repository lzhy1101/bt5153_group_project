### User Based Collaborative Filtering

```python
data=train_df
```

```python
# Preprocess
data[['user_id']]=data[['author.steamid']]
data['score']=np.where((data.recommended == True),1,-1)
```

User - Item matrix

```python
# Create dataframe for user based collaborative filtering recommendation
dataCF=data[['user_id','app_name','score']]
dataCF[['user_id']]=dataCF[['user_id']].astype(str)
dataCF[['app_name']]=dataCF[['app_name']].astype(str)
```

```python
# Creating a sparse pivot table with item in rows and user in columns
table=pd.pivot_table(dataCF,values  = 'score',index= 'app_name', columns = 'user_id')

# pearson correlation
corr=table.corr()
```

```python
table.shape
```

User - Attribute Matrix

```python
# Create User Attribute Dataframe
dataCB = data[[
    'user_id', 'review', 'steam_purchase', 'received_for_free',
    'author.num_games_owned', 'author.num_reviews', 'author.playtime_forever',
    'author.playtime_last_two_weeks', 'author.playtime_at_review',
    'author.last_played'
]]
```

```python
# Extract Sentiment from review
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
# Use negative portion as user attribute
dataCB['review_sentiment'] = dataCB['review'].apply(lambda x: analyzer.polarity_scores(x))
dataCB['review_sentiment_negportion']=dataCB['review_sentiment'].apply(lambda x: x['neg'])
```

```python
dataCB.head(3)
```

```python
# Data Pre-Processing

# Lable encoder categorial variable : steam_purchase,received_for_free,language
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder()

dataCB[['steam_purchase']]= label_encoder.fit_transform(dataCB[['steam_purchase']]) 
dataCB[['received_for_free']]=label_encoder.fit_transform(dataCB[['received_for_free']]) 
```

```python
# drop no use columns
dataCBF=dataCB.drop(columns=['review', 'review_sentiment'])
dataCBF['user_id']=dataCBF['user_id'].astype(str)
```

```python
# Standardize numerical values
scaler = preprocessing.StandardScaler()
dataCBF[[
    'author.num_games_owned', 'author.num_reviews', 'author.playtime_forever',
    'author.playtime_last_two_weeks', 'author.playtime_at_review',
    'author.last_played'
]] = scaler.fit_transform(dataCBF[[
    'author.num_games_owned', 'author.num_reviews', 'author.playtime_forever',
    'author.playtime_last_two_weeks', 'author.playtime_at_review',
    'author.last_played'
]])
```

```python
# summarize attributes for each users - take average of each attribute
table1=pd.pivot_table(dataCBF,aggfunc='mean',columns = 'user_id')
table1
```

```python
# Pearson Correlation Coefficient by user attribute matrix
corr2=table1.corr()
corr2
```

Combine User - Item matrix & User - Attribute Matrix

```python
# user-item correlation matrix fillna as 0 for calculation 
corr=corr.fillna(0)
# Combine two correlation matrix by assignment 0.8 weight to user-item matrix and 0.2 to user attribute matrix 
corr_combined=0.8*corr+0.2*corr2

corr_combined
```

Find 3 Similar Users

```python
user_id=list(data['user_id'].unique())
```

```python
# Create Similar users dataframe
similar_users=pd.DataFrame(columns=['user_id','similar_user1','similar_user2','similar_user3'])
similar_users[['user_id']]=pd.DataFrame(user_id)

for i in similar_users['user_id']:
    similar_users.loc[similar_users['user_id']==i,['similar_user1']]=corr_combined[str(i)].sort_values(ascending=False)[1:2].index[0]
    similar_users.loc[similar_users['user_id']==i,['similar_user2']]=corr_combined[str(i)].sort_values(ascending=False)[2:3].index[0]
    similar_users.loc[similar_users['user_id']==i,['similar_user3']]=corr_combined[str(i)].sort_values(ascending=False)[3:4].index[0]
```

Weighted Ratings and Recommendation

```python
# custom function to create unique set of web series
def unique_items():
    unique_items_list = []
    for i in table.columns:
        for j in table.index:
            unique_items_list.append(j)
    s=set(unique_items_list)
    unique_items_list=list(s)
    return unique_items_list

# Only recommend unplayed games
def recommend_games_to_users1(target_person):
    target_person_game_lst = []
    unique_list =unique_items()
    for i in table_original[str(target_person)].index:
        if pd.isna(table_original[str(target_person)].loc[i])==False:
            target_person_game_lst.append(i)
    s=set(unique_list)
    recommend_games=list(s.difference(target_person_game_lst))
    a = len(recommend_games)
    if a == 0:
        return 0
    return recommend_games

# score for each app based on similar users
# score* pearson correlation
def score(uid):
    numerator = table[similar_users.loc[similar_users['user_id'] == uid, ['similar_user1']].
      values[0][0]] * corr_combined.loc[
          similar_users.loc[similar_users['user_id'] == uid,
                            ['similar_user1']].values[0][0],
          [str(uid)]].values[0] + table[similar_users.loc[
              similar_users['user_id'] == uid,
              ['similar_user2']].values[0][0]] * corr_combined.loc[similar_users.loc[
                  similar_users['user_id'] == uid,
                  ['similar_user2']].values[0][0], [str(uid)]].values[0] + table[
                      similar_users.loc[similar_users['user_id'] == uid,
                                        ['similar_user3']].values[0]
                      [0]] * corr_combined.loc[similar_users.loc[
                          similar_users['user_id'] == uid,
                          ['similar_user3']].values[0][0], [str(uid)]].values[0]
    denominator = corr_combined.loc[
        similar_users.loc[similar_users['user_id'] == uid,
                          ['similar_user1']].values[0][0],
        [str(uid)]].values[0] + corr_combined.loc[
            similar_users.loc[similar_users['user_id'] == uid,
                              ['similar_user2']].values[0][0],
            [str(uid)]].values[0] + corr_combined.loc[similar_users.loc[
                similar_users['user_id'] == uid,
                ['similar_user3']].values[0][0], [str(uid)]].values[0]
    score=numerator/denominator
    score=pd.DataFrame(score)
    score=score.reset_index()
    score=score[score['app_name'].isin(recommend_games_to_users1(uid))]
    return score.sort_values(by=[0],ascending=False)
```

### Evaluation

```python
# Preprocess for testset
test_df[['user_id']]=test_df[['author.steamid']]
test_df['score']=np.where((test_df.recommended == True),1,-1)

# Create dataframe for user based collaborative filtering recommendation
test_df=test_df[['user_id','app_name','score']]
test_df[['user_id']]=test_df[['user_id']].astype(str)
test_df[['app_name']]=test_df[['app_name']].astype(str)
tabletest=pd.pivot_table(test_df,values  = 'score',index= 'app_name', columns = 'user_id')
```

```python
def true_playedlist(target_person):
    target_person_game_lst = []
    unique_list =unique_items()
    for i in tabletest[str(target_person)].index:
        if pd.isna(tabletest[str(target_person)].loc[i])==False:
            target_person_game_lst.append(i)
    return target_person_game_lst
```

```python
# Test set 
testList={}
for i in user_id:
    testList[i]=true_playedlist(i)
testList=pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in testList.items() ]))
# Total number of played games in test set
testcount=testList.count().sum()
```

```python
# Transfor to 2D array for further evaluation
testList_array=testList.to_numpy().T
recommender_array=recommender.to_numpy().T
recommender_array=recommender_array[:-1]
```

```python
import math
def indicators_5(rankedList, testList,testcount):
    #rankedLlist: 预测的二维数组
    #testList: test set的二维数组
    #将传入两个数组的推荐人对应上，比如说rankedlist的第一个是user a，那么testList的第一个也要是user a
    Hits_i = 0
    Len_R = 0
    Len_T = len(testList) # maximum number of games
    MRR_i = 0
    HR_i = 0
    HR_i2 = 0
    NDCG_i = 0    

    count=testcount
    for i in range(len(rankedList)): # i = recommender number of users
        for j in range(len(rankedList[i])): # j = recommender number of games
            for k in range(len(testList[i])): # k = test set number of games
                if testList[i][k]==rankedList[i][j]:
                    Hits_i+=1
                    HR_i+=1
                    HR_i2+=1
                    MRR_i+=1/(j+1)   
                    NDCG_i+=1/(math.log2(1+j+1))
                    break
    HR_i/=Len_T
    HR_i2/=count
    NDCG_i/=Hits_i
    print(Hits_i)
    print(f'HR_method1={HR_i}')
    print(f'HR_method2={HR_i2}')
    print(f'NDCG={NDCG_i}')
```
