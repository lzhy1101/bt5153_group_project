## 2 Spilt into train and test data

```python
size_dict = df.groupby(['author.steamid']).size().to_dict()
size_dict
```

```python
## train dataset 80% for each gamer by time
train_df=df.groupby(['author.steamid']).apply(lambda x : x.sort_values(by = 'datetime_created', ascending = True).head(round(len(x)*0.8))).reset_index(drop = True)
train_df[['author.steamid','datetime_created']].head(20)
```

```python
## test dataset 20% for each gamer by time
test_df=df.groupby(['author.steamid']).apply(lambda x : x.sort_values(by = 'datetime_created', ascending = True).tail(round(len(x)*0.2))).reset_index(drop = True)
test_df[['author.steamid','datetime_created']].head(20)
```

## 3 Evaluation Matrics

```python
test_df.head()
```

```python
# user_id
train_set_userids = set(train_df['author.steamid'].unique())
test_set_userids = set(test_df['author.steamid'].unique())
matching_users = train_set_userids.intersection(test_set_userids) 
len(matching_users)
```

```python
#### 为用户推荐N个游戏，N=user在test中的数量
#比如一个用户在test中有n条记录，则为其推荐最相关的n条
# filter掉每个用户已经出现在train中的推荐物品
# recommend后新增一列用于存放预测结果
# As a start we will ignore all the ratings where the actual value is not known. 
# Values with no known true rating cannot be used.
# 新增列名：pred_item_xx
recommend = pd.DataFrame(matching_users,columns=['user_id'])
recommend['pred']
recommend
```

```python
# 为每个用户计算hit ratio
hit_item = []
hit_ratio = []
for user_id in list(matching_users):
    # Find all the items user actually performed view/add/transact
    # test data中的物品
    true_item_ids = set(
        test_df[test_df['author.steamid'] == user_id]['app_id'].tolist())
    print(user_id, true_item_ids)
    # 为用户推荐的用品
    recommend_items = recommend[recommend['user_id'] == user_id]['pred']
    print(recommend_items)
    # Check if there there is any hit between user operations and recommendations
    hit = list(true_item_ids.intersection(set(recommend_items)))
    if hit:
        hit_ratio = len(hit) * 100 / (0.2*size_dict.get(user_id))
        hit_item.append(hit)
        hit_ratio.append(hit_ratio)
```
